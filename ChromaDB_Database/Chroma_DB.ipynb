```python
import chromadb

clinet = chromadb.Client()
collection = clinet.create_collection(name="Vector_store")
```

```python
collection.add(
    documents = [
        "This document is about New York",
        "This document is about New Delhi"
    ],
    ids = ["id1","id2"]
)
```

```python
collection.get()
```

```python
collection.get(ids=["id1"])
```

Now it is storing these things along with the context

```python
collection.query(
    query_texts = ['Query is about chole bhature'],
    n_results = 2
)
```

```python
collection.query(
    query_texts = ['Query is about Paris'],
    n_results = 2
)
```

```python
collection.add(
    documents = [
        "This document is about New York",
        "This document is about New Delhi"
    ],
    ids = ["id3","id4"],
    metadatas = [
        {"url": "https://en.wikipedia.org/wiki/New_York_City"},
        {"url": "https://en.wikipedia.org/wiki/New_Delhi"}
    ]
)
```

Now we have to include refrences in the data

```python
collection.query(
    query_texts = ['Query is about Paris'],
    n_results = 3
)
```
